2- Gráfico com porcentagem de matriculas ativas e inativas em relação ao modelos.

```python
import pandas as pd
import matplotlib.pyplot as plt


tabela = pd.read_csv('PDA_Cursos_Especializacao_Brasil.csv')

contagem_modalidades = tabela.groupby(['SITUACAO', 'MODALIDADE'])['VAGAS'].count()


plt.figure(figsize=(8, 6))
plt.pie(contagem_modalidades, labels=contagem_modalidades.index, autopct='%1.1f%%')
plt.title('Distribuição de Modalidades em Relação à Atividade e Inatividade das Matrículas')
plt.show()
```

5-Filtro vagas RJ por categoria.

```python

import pandas as pd

# Carregue os dados do arquivo CSV
arquivo = pd.read_csv("educacao.csv")

# Converta o DataFrame para df2
df2 = pd.DataFrame(arquivo)

# Prossiga com as operações necessárias
cidade = df2[df2['UF'] == 'RJ']
cursos_lins = cidade[["CATEGORIA_ADMINISTRATIVA", "QT_VAGAS_AUTORIZADAS"]].groupby("CATEGORIA_ADMINISTRATIVA").sum()

# Exiba os resultados
print(cursos_lins)
```

6-Gráfico repreentando quantidade de vagas disponíveis no estado do Rio de Janeiro em relação a Categoria.

```python
import matplotlib.pyplot as plt

categorias = ['Privada com fins lucrativos', 'Privada sem fins lucrativos', 'Pública Estadual', 'Pública Federal', 'Pública Municipal']
quantidades_rj = [150000, 50000, 10000, 20000, 30000] 

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(categorias, quantidades_rj, color='skyblue')
plt.xlabel('Categoria Administrativa')
plt.ylabel('Quantidade de Vagas Autorizadas')
plt.title('Quantidade de Vagas Autorizadas por Categoria no Rio de Janeiro')
plt.xticks(rotation=45)  
plt.tight_layout()


plt.show()
```

```python
import pandas as pd

# Carregue os dados do arquivo CSV
arquivo = pd.read_csv("educacao.csv")

# Converta o DataFrame para df2
df2 = pd.DataFrame(arquivo)

# Filtre os dados para o estado do Rio de Janeiro (UF == 'RJ')
cidade = df2[df2['UF'] == 'RJ']

# Agrupe por Categoria Administrativa e Modalidade de Ensino, somando as vagas autorizadas
cursos_lins = cidade.groupby(["CATEGORIA_ADMINISTRATIVA", "MODALIDADE"])["QT_VAGAS_AUTORIZADAS"].sum()

# Exiba os resultados
print(cursos_lins)
```
